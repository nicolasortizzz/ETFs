```python
import yfinance as yf
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

etfs = [
    "XLE", "XLF", "XLU", "XLI", "GDX", "XLK", "XLV", "XLY", "XLP", "XLB",
    "XOP", "IYR", "XHB", "ITB", "VNQ", "GDXJ", "IYE", "OIH", "XME", "XRT",
    "SMH", "IBB", "KBE", "KRE", "XTL"
]

# Descargar precios ajustados desde 2014 (auto_adjust=True por defecto)
data = yf.download(etfs, start="2014-01-01", end="2024-01-01")["Close"]

# Normalizar (base 1)
data_normalizada = data / data.iloc[0]

# Graficar
plt.figure(figsize=(18, 10))
for etf in etfs:
    plt.plot(data_normalizada[etf], label=etf)

plt.title("Evolución de ETFs Sectoriales (2014–2024)")
plt.xlabel("Fecha")
plt.ylabel("Evolución (Base = 1)")
plt.legend(loc="upper left", fontsize=8, ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
try:
    from tabulate import tabulate
except ImportError:
    tabulate = None

etfs = [
    'XLE', 'XLF', 'XLU', 'XLI', 'GDX', 'XLK', 'XLV', 'XLY', 'XLP', 'XLB', 
    'XOP', 'IYR', 'XHB', 'ITB', 'VNQ', 'GDXJ', 'IYE', 'OIH', 'XME', 'XRT', 
    'SMH', 'IBB', 'KBE', 'KRE', 'XTL'
]

start_date = '2010-01-01'
end_date = '2025-12-31'

# Descargar datos de cierre ajustado
data = yf.download(etfs, start=start_date, end=end_date, group_by='ticker', auto_adjust=True)

# Extraer precios de cierre
if isinstance(data.columns, pd.MultiIndex):
    close_prices = data.xs('Close', axis=1, level=1)
else:
    close_prices = data[['Close']]

# Eliminar ETFs sin datos
close_prices = close_prices.dropna(axis=1, how='all')

# Calcular el retorno total de cada ETF en el periodo
growth = (close_prices.iloc[-1] / close_prices.iloc[0]) - 1

# Calcular desvío estándar de los retornos diarios
returns = close_prices.pct_change().dropna()
std_devs = returns.std()

# Crear un DataFrame con los resultados
result = pd.DataFrame({
    'Crecimiento (%)': growth * 100,
    'Desvío estándar': std_devs
})

# Ordenar por crecimiento descendente y mostrar top 10
top_10 = result.sort_values(by='Crecimiento (%)', ascending=False).head(10)
top_10 = top_10.round(2)

# --- Salida bonita en consola ---
print("\n" + "="*60)
print("Top 10 ETFs que más crecieron (2010-2025):\n")
if tabulate:
    print(tabulate(top_10[['Crecimiento (%)']], headers='keys', tablefmt='psql', showindex=True))
else:
    print(top_10[['Crecimiento (%)']])

# --- Visualización solo con variación ---
plt.figure(figsize=(12,6))
bars = plt.barh(top_10.index, top_10['Crecimiento (%)'], color=sns.color_palette("viridis", len(top_10)))

plt.gca().set_ylabel("")
plt.yticks(ticks=range(len(top_10.index)), labels=top_10.index, fontsize=11)
plt.xlabel('Crecimiento total (%)')
plt.title('Top 10 ETFs que más crecieron (2010-2025)')

# Quitar todos los bordes (spines)
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Solo la variación fuera de la barra, bien posicionada
for i, (etf, row) in enumerate(top_10.iterrows()):
    val = row['Crecimiento (%)']
    x_pos = val + max(top_10['Crecimiento (%)']) * 0.015
    plt.text(x_pos, i, f"{val:.2f}%", va='center', ha='left', fontsize=12, fontweight='bold')

plt.grid(axis='x', linestyle=':', alpha=0.7)
plt.tight_layout()
plt.show()
```

### **Sectores con mayor crecimiento**

- El liderazgo de tecnología y semiconductores fue innegable, obteniendo rendimientos con un amplia diferencia a los demas sectores. Si bien la industria de los semiconductores ya era lider en 2015, a partir de esta fecha se nota un efecto diferencial que la posiciono como el sector que mas crecio con diferencia. Cabe destacar que los semiconductores tienen una influencia total en la inteligencia artificial por lo que se espera que el sector siga creciendo. 
- El sector del Real State tambien demostro un gran crecimiento. Sin contradecir la historia este sector sigue siendo fuertemente pugante y se espera que lo siga siendo.
- Sectores clasicos como el consumo minorista, servicios financieros y el sector industrial siguen liderando. 
- La aparicion de la biotecnologia como un de los sectores que mas crecio es una novedad historicamente hablando.

```python

etfs = [
    'XLE', 'XLF', 'XLU', 'XLI', 'GDX', 'XLK', 'XLV', 'XLY', 'XLP', 'XLB', 
    'XOP', 'IYR', 'XHB', 'ITB', 'VNQ', 'GDXJ', 'IYE', 'OIH', 'XME', 'XRT', 
    'SMH', 'IBB', 'KBE', 'KRE', 'XTL'
]

start_date = '2010-01-01'
end_date = '2025-12-31'

# Descarga los datos
data = yf.download(etfs, start=start_date, end=end_date, group_by='ticker', auto_adjust=True)

# Revisamos si hay MultiIndex y obtenemos correctamente la columna 'Close'
if isinstance(data.columns, pd.MultiIndex):
    # Columnas tipo: (ticker, field) -> tomamos solo los 'Close'
    close_prices = data.xs('Close', axis=1, level=1)
else:
    # Un solo ticker, columnas planas
    close_prices = data[['Close']]

# Eliminamos columnas (tickers) sin datos
close_prices = close_prices.dropna(axis=1, how='all')

# Calculamos retornos
returns = close_prices.pct_change().dropna()

# Calculamos la matriz de correlación
correlation_returns = returns.corr()

# Mostramos la matriz por consola
print("Matriz de correlaciones:")
print(correlation_returns)

# Graficamos el heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_returns, annot=True, fmt=".2f", cmap='coolwarm', linewidths=.5)
plt.title('Correlación de los retornos diarios de los ETFs seleccionados desde 2010')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

etfs = [
    'SMH', 'XLK', 'XLY', 'ITB', 'XHB', 
    'XLI', 'XLF', 'XLV', 'XRT', 'IBB'
]

start_date = '2010-01-01'
end_date = '2025-12-31'

# Descargar datos de cierre ajustado
data = yf.download(etfs, start=start_date, end=end_date, group_by='ticker', auto_adjust=True)

# Extraer precios de cierre
if isinstance(data.columns, pd.MultiIndex):
    close_prices = data.xs('Close', axis=1, level=1)
else:
    close_prices = data[['Close']]

# Eliminar columnas vacías
close_prices = close_prices.dropna(axis=1, how='all')

# Calcular retornos diarios
returns = close_prices.pct_change().dropna()

# Matriz de correlación
corr_matrix = returns.corr()

print("\nMatriz de correlación entre los 10 ETFs:\n")
print(corr_matrix.round(2))

# Obtener las correlaciones más bajas (excluyendo duplicados y la diagonal)
corr_pairs = corr_matrix.where(~np.eye(corr_matrix.shape[0], dtype=bool)).stack()
corr_pairs.index = corr_pairs.index.rename(['ETF1', 'ETF2'])  # Nombres claros

# Evitar duplicados (pares inversos)
corr_pairs = corr_pairs.reset_index()
corr_pairs.columns = ['ETF1', 'ETF2', 'Correlacion']
corr_pairs = corr_pairs[corr_pairs['ETF1'] < corr_pairs['ETF2']]

# Las 5 correlaciones más bajas
lowest_corr = corr_pairs.nsmallest(10, 'Correlacion')

print("\nLas 10 correlaciones más bajas entre estos ETFs:\n")
for i, row in lowest_corr.iterrows():
    print(f"{row['ETF1']} - {row['ETF2']}: {row['Correlacion']:.3f}")
```

### **Bajas correlaciones destacables**
- Los sectores de biotecnologia y medicina estan bajamente correlacionados con los demas sectores, pero muestran estar correlacionados entre si. 
- El sector de semiconductores muestra baja correlacion con el Real State, volviendo al Real State un opcion atractiva para combinar junto con empresas de semiconducotres.
